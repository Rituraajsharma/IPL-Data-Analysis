# Q.1 Total Runs Scored by Each Team:

```python
runs_per_team = df.groupby('batting_team')['total_runs'].sum()
print(runs_per_team)
```

```python
plt.figure(figsize=(10,6))
plt.bar(runs_per_team.index, runs_per_team.values)
plt.xlabel("Team")
plt.ylabel("Total Runs")
plt.title("Total Runs Scored by Each Team")
plt.xticks(rotation=90)
plt.show()
```

# Q.2 Top 10 Batsmen with Highest Runs

```python
top_batsmen = df.groupby('batter')['batsman_runs'].sum().nlargest(10)
print(top_batsmen)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_batsmen.index, top_batsmen.values, color='skyblue')
plt.xlabel("Batsman")
plt.ylabel("Total Runs")
plt.title("Top 10 Batsmen with Highest Runs")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#Q.3 Dismissal Type Distribution

```python
dismissal_type_counts = df['dismissal_kind'].value_counts()
print(dismissal_type_counts)
```

```python
plt.figure(figsize=(8, 10))
plt.pie(dismissal_type_counts.values, labels=dismissal_type_counts.index, autopct='%1.1f%%')
plt.title("Dismissal Type Distribution")
plt.show()
```

# Q.4 Which bowling team has conceded the most extra runs in the IPL?

```python
extra_runs_by_team = df.groupby('bowling_team')['extra_runs'].sum().sort_values(ascending=False)
print(extra_runs_by_team)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(extra_runs_by_team.index, extra_runs_by_team.values, color='orange')
plt.xlabel("Bowling Team")
plt.ylabel("Total Extra Runs Conceded")
plt.title("Extra Runs Conceded by Bowling Teams")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#Q.5 Total Runs Scored by Each Team

# Q.6 Top 10 Bowlers with Most Wickets

```python
top_bowlers = df[df['dismissal_kind'] != 'run out'].groupby('bowler')['dismissal_kind'].count().nlargest(10)
print(top_bowlers)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_bowlers.index, top_bowlers.values, color='green')
plt.xlabel("Bowler")
plt.ylabel("Total Wickets")
plt.title("Top 10 Bowlers with Most Wickets")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Q.7 Number of Sixes Hit by Each Team

```python
sixes_per_team = df[df['batsman_runs'] == 6].groupby('batting_team')['batsman_runs'].count()
print(sixes_per_team)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sixes_per_team.index, sixes_per_team.values, color='red')
plt.xlabel("Team")
plt.ylabel("Number of Sixes")
plt.title("Number of Sixes Hit by Each Team")
plt.xticks(rotation=90)
plt.show()
```

#Q.8 Most Successful Chasing Team

```python
chasing_wins = df[df['inning'] == 2].groupby('batting_team')['match_id'].nunique().sort_values(ascending=False)
print(chasing_wins)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(chasing_wins.index, chasing_wins.values, color='purple')
plt.xlabel("Team")
plt.ylabel("Number of Successful Chases")
plt.title("Most Successful Chasing Team")
plt.xticks(rotation=90)
plt.show()
```

#Q.9 Most Successful Defending Team

```python
defending_wins = df[df['inning'] == 1].groupby('batting_team')['match_id'].nunique().sort_values(ascending=False)
print(defending_wins)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(defending_wins.index, defending_wins.values, color='brown')
plt.xlabel("Team")
plt.ylabel("Number of Successful Defenses")
plt.title("Most Successful Defending Team")
plt.xticks(rotation=90)
plt.show()
```

#Q.10 Most Wickets Taken in Death Overs (Overs 16-20)

```python
death_overs_wickets = df[df['over'].between(16, 20) & (df['dismissal_kind'] != 'run out')].groupby('bowler')['dismissal_kind'].count().nlargest(10)
print(death_overs_wickets)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(death_overs_wickets.index, death_overs_wickets.values, color='magenta')
plt.xlabel("Bowler")
plt.ylabel("Total Wickets in Death Overs")
plt.title("Most Wickets Taken in Death Overs")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
